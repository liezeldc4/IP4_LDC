### 5.	Remove all the $ symbols from the 'Price' column and convert  and update it to  64-bit int numbers using astype(int64) Hint. Look for string functions like replace/strip).

```python
car_sales['Price'] = car_sales['Price'].str.replace('$','').str.replace(',','').astype(np.int64)
```

```python
car_sales.head()
```

```python
car_sales.info() # Here we can see price has been converted from object to int64
```

### 6.	Add a column called 'Total Sales' to car_sales which cumulatively adds the 'Price' column look up the cumsum() function.

```python
car_sales['Total Sales'] = np.cumsum(car_sales['Price'])
car_sales.head()
```

### 7.	Add a column called 'Sale Date' which lists a series of successive dates starting from today (your today). Check out date_range function and keep the periods=len(car_sales).

```python
car_sales['Sale Date'] = pd.date_range(start = pd.datetime.now().date(),
                                      end = None,
                                      periods = len(car_sales))
```

```python
car_sales.head()
```

### 8.	Use the matplotlib line plot function to plot the 'Price' column versus the 'Total Sales' column (Give proper title and appropriate labels).

```python
plt.plot(car_sales['Price'],
        car_sales['Total Sales'])

plt.xlabel('Price')
plt.ylabel('Total Sales')
plt.title('Price vs Total Sales Line Graph')

plt.show()
```

### 9.	Create a matplotlib scatter plot of the 'Odometer (KM)' vs 'Price' column.

```python
plt.scatter(car_sales['Odometer (KM)'],
           car_sales['Price'])

plt.xlabel('Odometer (KM)')
plt.ylabel('Price')
plt.title('Odometer (KM) vs Price Scatter Graph')

plt.show()
```

### 10.	Create a matplotlib bar graph of the 'Make' vs 'Odometer (KM)' columns

```python
plt.bar(car_sales['Make'],
       car_sales['Odometer (KM)'])

plt.xlabel('Make')
plt.ylabel('Odometer (KM)')
plt.title('Make vs Odometer(KM) Bar Graph')

plt.show()
```

### 11.	Create a histogram of the 'Make' column

```python
plt.hist(car_sales['Make'])
plt.title('Histogram Of Car Make')
plt.xlabel('Make')
plt.show()
```

### 12.	Create a subplot of multiple plots (step 8,9,10,11).

```python
fig = plt.figure(figsize = (12,12))

axis1 = fig.add_subplot(221)

axis1.plot(car_sales['Price'],
        car_sales['Total Sales'])

plt.xlabel('Price')
plt.ylabel('Total Sales')
plt.title('Price vs Total Sales Line Graph')

axis2 = fig.add_subplot(222)

axis2.scatter(car_sales['Odometer (KM)'],
           car_sales['Price'])

plt.xlabel('Odometer (KM)')
plt.ylabel('Price')
plt.title('Odometer (KM) vs Price Scatter Graph')

axis3 = fig.add_subplot(223)

axis3.bar(car_sales['Make'],
       car_sales['Odometer (KM)'])

plt.xlabel('Make')
plt.ylabel('Odometer (KM)')
plt.title('Make vs Odometer(KM) Bar Graph')

axis4 = fig.add_subplot(224)

axis4.hist(car_sales['Make'])
plt.title('Histogram Of Car Make')
plt.xlabel('Make')
plt.show()
```

### 13.	Create a box plot of Colour and Total Sales using seaborn plot.

```python
sns.boxplot(x = car_sales['Colour'], 
            y = car_sales['Total Sales'])
plt.show()
```

### 14.	Find out the relationship between "Total Sales" and "Odometer (KM)" using linear regression with no error bar and car make will determine the hue

```python
sns.lmplot(x='Total Sales',
           y='Odometer (KM)',
           data = car_sales,
           ci=False,
           hue = 'Make')
plt.show()
```
